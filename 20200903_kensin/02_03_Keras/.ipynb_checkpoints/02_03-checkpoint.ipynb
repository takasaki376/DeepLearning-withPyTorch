```python
import numpy as np
# 訓練データと検証データの分割
from sklearn.model_selection import train_test_split
# 予測結果の精度確認（ROC曲線の描画）
from sklearn.metrics import roc_auc_score, roc_curve
# データを扱う
import pandas as pd
# グラフ描画
from matplotlib import pyplot as plt
```

```python
# モデル
from keras.models import Sequential
# 層の指定（全結合）
from keras.layers import Dense
# 活性化関数
from keras.layers import Activation
```

## 訓練データの読み込み

```python
# CSVファイル読み込み（訓練データ）
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
# 性別を変換する（訓練データ）
df_train["Gender"] = df_train["Gender"].apply(lambda x: 1 if x=="Male" else 0)
df_train.head()
```

## 評価データの読み込み

```python
# CSVファイル読み込み（評価データ）
df_vat = pd.read_csv('test.csv')
df_vat.head()
```

```python
# 性別を変換する（評価データ）
df_vat["Gender"] = df_vat["Gender"].apply(lambda x: 1 if x=="Male" else 0)
df_vat.head()
```

## データを加工する

```python
def xConv(x_tmp):
    # T_Bil、D_Bilの変換
    bins1 = [0, 0.6,  0.7, 0.75, 0.8, 0.9, 1.0, 1.2, 1.3, 1.4, 1.5, 1.8, 100]
    bins2 = [0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.4, 0.5, 0.6, 0.8, 100]
    x_tmp=comCutCnnect(x_tmp, bins1, 'T_Bil' , bins2, 'D_Bil' )
    
    # T_Bil、D_Bilの変換
    bins1 = [0, 5, 7.5, 10, 12.5, 15, 17.5, 20, 22.5, 25, 50, 100, 125,1500]
    bins2 = [0, 20, 25,  30, 35,  40, 50, 60, 70, 80, 100, 140, 180,900]
    x_tmp=comCutCnnect(x_tmp, bins1, 'ALT_GPT' , bins2, 'AST_GOT' )
    
    return x_tmp


def comCutCnnect(x_tmp,bins1 ,col1, bins2 ,col2 ):   
    X_cut_1 = pd.cut(x_tmp[col1], bins=bins1, labels=False)
    X_cut_2 = pd.cut(x_tmp[col2], bins=bins2, labels=False)
    X_cut = pd.DataFrame(np.array(X_cut_1) * np.array(X_cut_2) , columns={col1})
    x_tmp.drop(columns=[col1 , col2], inplace=True, axis=1)
    x_tmp = pd.concat([x_tmp, X_cut], axis=1)
    return x_tmp
```

```python
# 説明変数（訓練データ）
x_tmp=xConv(df_train.loc[:,['Age', 'Gender', 'T_Bil', 'D_Bil', 'ALP', 'ALT_GPT', 'AST_GOT', 'TP', 'Alb','AG_ratio']])
x_train = np.array(x_tmp)
print(x_tmp.head())
# 目的変数（訓練データ）
y_train = np.array(df_train.loc[:,['disease']])


# 説明変数（評価データ）
x_tmp=xConv(df_vat)
x_vat = np.array(x_tmp)
print(x_tmp.head())
```

```python
y = np.zeros((len(y_train), 1 + y_train.max()), dtype=int)
y[np.arange(len(y_train)), y_train.T] = 1
y_train = y
X_train,X_test,Y_train,Y_test=train_test_split(x_train, y_train, test_size=0.25, random_state=1) 
```

```python
# 行数、列数の確認
in_size=x_train.shape[1]
print(x_train.shape)
```

```python
out_size=y_train.shape[1]
y_train.shape
```

# モデルの定義

```python
node1=40
node2=15
node3=8

# 以下、ネットワークを構築
model = Sequential()
# 第1層
model.add(Dense(units=node1, input_shape  = (in_size , )))
model.add(Activation("relu"))
# 第2層
model.add(Dense(node2))
model.add(Activation("relu"))
# 第3層
model.add(Dense(node3))
model.add(Activation("relu"))
# 出力層
model.add(Dense(out_size))
model.add(Activation("softmax"))
```

```python
# モデルの確認
model.summary()
```

```python
# パラメータの設定 
#  mean_squared_error   平均二乗誤差
#  categorical_crossentropy   クロスエントロピー
#model.compile(loss="categorical_crossentropy",
#             optimizer='adam',
#             metrics=["accuracy"])
model.compile(loss="mean_squared_error",
             optimizer='adam',
             metrics=["accuracy"])
```

```python
result = model.fit(x=X_train, y=Y_train, epochs=200, batch_size=15)
```

## 学習曲線（損失）の確認

```python
# 折れ線グラフによる学習データの損失の描画
plt.plot(result.history['loss'])

# 凡例の指定
plt.legend(['Train', 'Val'])

# グラフの軸タイトルの指定
plt.xlabel('Epoch')
plt.ylabel('Loss')

# 描画の実行
plt.show()
```

## 検証データを使って予測値の検証

```python
# 検証 <-訓練データ 
score = model.evaluate(X_train, Y_train, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# 検証 <-訓練データ 
score = model.evaluate(X_test, Y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
def y_conv(y):
    tmp = []
    for idx in y:
        if idx[0]  <  idx[1]:
            tmp.append(1)
        else:
            tmp.append(0)
    return tmp
```

```python
y_pred = model.predict(X_test)  
y_pred = y_conv(y_pred)
y_true = y_conv(Y_test)
fpr, tpr, thresholds = roc_curve(y_true=y_true, y_score=y_pred)
plt.plot(fpr, tpr, label='roc curve')
plt.plot([0, 1], [0, 1], linestyle=':', label='random')
plt.plot([0, 0, 1], [0, 1, 1], linestyle=':', label='ideal')
plt.legend()
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.show()
```

```python
# AUCスコアの算出
auc_score = roc_auc_score(y_true=y_true, y_score=y_pred)
print("AUC:", auc_score)
```

#  予測する

```python
y_predict = model.predict(x_vat)  
y_predict = y_conv(y_predict)
df_out = pd.DataFrame(y_predict, columns=['val'])

df_out.to_csv('./submit.csv', encoding='utf_8_sig' , header=False)
```
