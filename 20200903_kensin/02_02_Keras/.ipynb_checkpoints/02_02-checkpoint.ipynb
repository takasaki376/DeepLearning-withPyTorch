```python
import numpy as np
# 訓練データと検証データの分割
from sklearn.model_selection import train_test_split
# データを扱う
import pandas as pd
# グラフ描画
from matplotlib import pyplot as plt
```

```python
# モデル
from keras.models import Sequential
# 層の指定（全結合）
from keras.layers import Dense
# 活性化関数
from keras.layers import Activation
```

## 訓練データの読み込み

```python
# CSVファイル読み込み（訓練データ）
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
# 性別を変換する（訓練データ）
df_train["Gender"] = df_train["Gender"].apply(lambda x: 1 if x=="Male" else 0)
df_train.head()
```

## 評価データの読み込み

```python
# CSVファイル読み込み（評価データ）
df_vat = pd.read_csv('test.csv')
df_vat.head()
```

```python
# 性別を変換する（評価データ）
df_vat["Gender"] = df_vat["Gender"].apply(lambda x: 1 if x=="Male" else 0)
df_vat.head()
```

```python
# DataFrameからscatter_matrixを作成
from pandas.plotting import scatter_matrix
import mglearn
x=df_train.loc[:,['Age', 'Gender', 'T_Bil', 'D_Bil', 'ALP', 'ALT_GPT', 'AST_GOT', 'TP', 'Alb','AG_ratio']]
y=y_train = np.array(df_train.loc[:,['disease']])
grr = scatter_matrix(x, c=y,  figsize=(20, 20) , cmap=mglearn.cm3)
plt.show()
```

```python
x_tmp=df_train.loc[:,['Age', 'Gender', 'T_Bil', 'D_Bil', 'ALP', 'ALT_GPT', 'AST_GOT', 'TP', 'Alb','AG_ratio']]
x_tmp.where(x_tmp['T_Bil']<1.8 , 1.8 , inplace=True)
x_tmp.where(x_tmp['D_Bil']<0.8 , 0.8 , inplace=True)
# 説明変数（訓練データ）
x_train = np.array(x_tmp)
# 目的変数（訓練データ）
y_train = np.array(df_train.loc[:,['disease']])


x_tmp=df_vat
x_tmp.where(x_tmp['T_Bil']<1.8 , 1.8 , inplace=True)
x_tmp.where(x_tmp['D_Bil']<0.8 , 0.8 , inplace=True)
# 説明変数（評価データ）
x_test = np.array(x_tmp)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_train, y_train, test_size=0.25, random_state=1) 
```

```python
# 行数、列数の確認
in_size=x_train.shape[1]
print(x_train.shape)
```

```python
out_size=y_train.shape[1]
y_train.shape
```

# モデルの定義

```python
node1=28
node2=5

# 以下、ネットワークを構築
model = Sequential()
# 第1層
model.add(Dense(units=node1, input_shape  = (in_size , )))
model.add(Activation("relu"))
# 第2層
model.add(Dense(node2))
model.add(Activation("relu"))
# 出力層
model.add(Dense(out_size))
model.add(Activation("softmax"))
```

```python
# モデルの確認
model.summary()
```

```python
# パラメータの設定 
#  mean_squared_error   平均二乗誤差
#  categorical_crossentropy   クロスエントロピー
model.compile(loss="categorical_crossentropy",
             optimizer='adam',
             metrics=["accuracy"])
```

```python
result = model.fit(x=X_train, y=Y_train, epochs=80, batch_size=24)
```

## 学習曲線（損失）の確認

```python
# 折れ線グラフによる学習データの損失の描画
plt.plot(result.history['loss'])

# 凡例の指定
plt.legend(['Train', 'Val'])

# グラフの軸タイトルの指定
plt.xlabel('Epoch')
plt.ylabel('Loss')

# 描画の実行
plt.show()
```

## 検証データを使って予測値の検証

```python
# 検証 <-訓練データ 
score = model.evaluate(X_train, Y_train, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# 検証 <-訓練データ 
score = model.evaluate(X_test, Y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

#  予測する

```python
y_predict = model.predict(x_test)       
df_out = pd.DataFrame(y_predict, columns=['val'])

df_out.to_csv('./submit.csv', encoding='utf_8_sig' , header=False)
```
