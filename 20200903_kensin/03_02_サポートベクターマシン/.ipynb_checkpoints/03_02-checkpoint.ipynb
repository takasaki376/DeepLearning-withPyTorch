```python
import numpy as np
#サイキットラーン　（メモリの節約のため、sklearnの中からsvmのみをインポートする
from sklearn import svm
# 訓練データと検証データの分割
from sklearn.model_selection import train_test_split
# 予測結果の精度表示
from sklearn import metrics
# データを扱う
import pandas as pd
# グラフ描画
from matplotlib import pyplot as plt
import japanize_matplotlib
```

## 訓練データの読み込み

```python
# CSVファイル読み込み（訓練データ）
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
# 性別を変換する（訓練データ）
df_train["Gender"] = df_train["Gender"].apply(lambda x: 1 if x=="Male" else 0)
df_train.head()
```

## 評価データの読み込み

```python
# CSVファイル読み込み（評価データ）
df_vat = pd.read_csv('test.csv')
df_vat.head()
```

```python
# 性別を変換する（評価データ）
df_vat["Gender"] = df_vat["Gender"].apply(lambda x: 1 if x=="Male" else 0)
df_vat.head()
```

## 訓練

```python
x_tmp=df_train.loc[:,['Age', 'Gender', 'T_Bil', 'D_Bil', 'ALP', 'ALT_GPT', 'AST_GOT', 'TP', 'Alb','AG_ratio']]
x_tmp.where(x_tmp['T_Bil']<1.8 , 1.8 , inplace=True)
x_tmp.where(x_tmp['D_Bil']<0.8 , 0.8 , inplace=True)
# 説明変数（訓練データ）
x_train = np.array(x_tmp)
# 目的変数（訓練データ）
y_train = np.array(df_train.loc[:,['disease']])


x_tmp=df_vat
x_tmp.where(x_tmp['T_Bil']<1.8 , 1.8 , inplace=True)
x_tmp.where(x_tmp['D_Bil']<0.8 , 0.8 , inplace=True)
# 説明変数（評価データ）
x_test = np.array(x_tmp)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_train, y_train, test_size=0.25, random_state=1) 
```

```python
# サポートベクターマシーン
clf = svm.SVC()  
# 訓練
clf.fit(X_train, Y_train)  
```

## 検証データを使って予測値の検証

```python
# 検証 <-訓練データ 
clf.score(X_train ,Y_train)
```

```python
# 検証 <-検証データ
# (検証データと値が近いなら信用できる。
#  どこまでの差異を近いとするかはそれぞれ
#  判断する必要がある。)
clf.score(X_test,Y_test)
```

## 予測する

```python
y_predict = clf.predict(x_test)       
df_out = pd.DataFrame(y_predict, columns=['val'])
df_out.to_csv('./submit.csv', encoding='utf_8_sig' , header=False)
```
